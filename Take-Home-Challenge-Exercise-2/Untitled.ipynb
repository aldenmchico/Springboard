```python
import pandas as pd
import numpy as np

from datetime import datetime as dt
```

Goal: Identify which factors predict future user adoption. <i><b>An adopted user is a user who has logged into the product on three separate days in at least one seven-day period. </b></i>

```python
# Import takehome_users.csv
user_info_df = pd.read_csv('takehome_users.csv', encoding='latin-1')
user_info_df.head()
```

```python
# Clean the df for relevant information
user_info_df = user_info_df.drop(columns=['name', 'email'])
user_info_df = user_info_df.fillna(0)
user_info_df['creation_time'] = pd.to_datetime(user_info_df['creation_time'])
user_info_df.head()
```

```python
user_engagement_df = pd.read_csv('takehome_user_engagement.csv')
user_engagement_df.head()
```

```python
user_engagement_df = user_engagement_df.sort_values(by=['user_id', 'time_stamp'])
user_engagement_df.head()
```

```python
count_engagement_df = user_engagement_df.groupby('user_id').count()
count_engagement_df = count_engagement_df.loc[count_engagement_df.visited >= 3]
count_engagement_df.head()
```

```python
user_engagement_df = pd.merge(user_engagement_df, count_engagement_df, on='user_id', how='right')
user_engagement_df.head()
```

```python
# Import takehome_users.csv
example = user_engagement_df.pivot_table(index='user_id', values='time_stamp_x')
example.head()
```

```python
adopt_user_id = user_engagement_df.user_id.unique()
adopt_user_id
```

```python
adopt_user_info_df = user_info_df.loc[user_info_df.object_id.isin(adopt_user_id), :]
adopt_user_info_df = adopt_user_info_df.sort_values('object_id')
adopt_user_info_df.head()
```
